## Train Test Split

```python
X= df_Sample.drop(['Class'], axis= 1)
y= df_Sample['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Model Building

## Logistic Regression

```python
#Instantiate logistic reg
from sklearn.linear_model import LogisticRegression
log_Reg= LogisticRegression()
```

```python
#fit the data
log_Reg.fit(X_train, y_train)
```

```python
#prediction
log_Reg_pred= log_Reg.predict(X_test)
```

```python
#evaluation metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print("Accuracy Score: ", accuracy_score(y_test,log_Reg_pred))
print("Classification Report :")
print(classification_report(y_test,log_Reg_pred))
```

```python
#forming a dataframe for actual and predicted values with XGBoost
print("actual and predicted values for training dataset: ")
pd.DataFrame({'Actual' : y_test, 'Predicted' :log_Reg_pred})
```

```python
#erros found by models 
n_errors = (log_Reg_pred != y_test).sum()
```

```python
print("Errors detected from logistic Regression", n_errors)
```
